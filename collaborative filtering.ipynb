```python
ratings_with_name = ratings.merge(anime,on='anime_id')
```

```python
x = ratings_with_name.groupby('user_id').count()['rating_y'] > 250
otaku = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['user_id'].isin(otaku)]
```

```python
y = filtered_rating.groupby('name').count()['rating_y']>=50
famous_anime = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['name'].isin(famous_anime)]
```
